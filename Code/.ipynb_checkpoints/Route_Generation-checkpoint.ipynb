#### Create transit networks, one for each day of the week
Here, we create a travel time weighted graph from the GTFS feed.
A network is created for Friday and Saturday between 12:00 AM and 11:59 PM to represent travel times throughout an entire day.
What results is a global urbanaccess_network object that holds a graph comprised of nods and edges for the processed GTFS network data for transit services operating at the specified days/times inside of transit_edges and transit_nodes.

```python
transit_net_friday = ua.gtfs.network.create_transit_net(
    gtfsfeeds_dfs=loaded_feeds,
    day='friday',
    timerange=['00:00:00', '23:59:59'])
```

```python
transit_net_saturday = ua.gtfs.network.create_transit_net(
    gtfsfeeds_dfs=loaded_feeds,
    day='saturday',
    timerange=['00:00:00', '23:59:59'])
```

```python
transit_net_friday.transit_nodes.plot(kind='scatter', x='x', y='y', s=0.1)
```

```python
transit_net_saturday.transit_nodes.plot(kind='scatter', x='x', y='y', s=0.1)
```

#### Download OSM Pedestrian Street Network Data

```python
osm_nodes, osm_edges = ua.osm.load.ua_network_from_bbox(bbox=bbox,
                                                        remove_lcn=True)
```

#### Create the pedestrian network.
Create a travel time weighted graph from the retrieved OSM pedestrian network and add to the existing UrbanAccess network object. Assume that a pedestrian travels 3 mph.

```python
osm_network = ua.osm.network.create_osm_net(
    osm_edges=osm_edges,
    osm_nodes=osm_nodes,
    travel_speed_mph=3,
    network_type='walk')

# osm_network_bike = ua.osm.network.create_osm_net(
#     osm_edges=osm_edges,
#     osm_nodes=osm_nodes,
#     travel_speed_mph=12,
#     network_type='bike')
```

#### Add average headways to network travel time

```python
loaded_feeds = ua.gtfs.headways.headways(loaded_feeds,
                                         headway_timerange=['00:00:00', '23:59:59'])
```

#### Combine the transit and pedestrian networks

```python
access_nw_walk = ua.network.integrate_network(
    urbanaccess_network=osm_network,
    headways=True,
    urbanaccess_gtfsfeeds_df=loaded_feeds,
    headway_statistic='mean')
```

```python
access_nw_bike = ua.network.integrate_network(
    urbanaccess_network=osm_network,
    headways=True,
    urbanaccess_gtfsfeeds_df=loaded_feeds,
    headway_statistic='mean')
```

#### Save the network to disk

```python
ua.network.save_network(urbanaccess_network=access_nw_walk,
                        filename='Access_Network_Walk.h5',
                        overwrite_key = True)
```

# Guidance from https://github.com/UDST/pandana/blob/3e3d35ca2d57428714b89ed8fc7020bc55067e1d/examples/Pandana-demo.ipynb

#### Initialize the Pandana network

```python
# The from_int, to_int and twoway=False parameters denote this is an explicit one-way network
s_time = time.time()
transit_ped_net = pdna.Network(access_nw.net_nodes["x"],
                               access_nw.net_nodes["y"],
                               access_nw.net_edges["from_int"],
                               access_nw.net_edges["to_int"],
                               access_nw.net_edges[["weight"]], 
                               twoway=False)
print('Took {:,.2f} seconds'.format(time.time() - s_time))
```

```python
access_nw.net_nodes
```

```python
# Loading trajectories from CSV
traj = pd.read_csv("../Data/Scenarios_Synthetic_Data.csv")

# Split latlon into separate columns
traj[["latitude","longitude"]] = traj["latlon"].str.split(',', 1, expand=True)
# Set data types for both latitude and longitude columns as float64
traj["latitude"] = traj["latitude"].astype('float64')
traj["longitude"] = traj["longitude"].astype('float64')
# Set index as the time
traj["datetime"] = pd.to_datetime(traj["time"])#, format='%m/%d/%Y %H:%M')
traj.set_index('datetime')

traj
```

```python
# Generate new offset column of latlon
traj['latlon_shift'] = traj.groupby('person_id')['latlon'].shift(-1)

# Identify the nearest node to each stop of individuals' trajectories
traj['node_id_o'] = transit_ped_net.get_node_ids(traj.longitude, traj.latitude)
# Set a placeholder of the same nearest nodes for the node id for the next location (latlon_shift)
traj['node_id_d'] = transit_ped_net.get_node_ids(traj.longitude, traj.latitude)

# Filter rows where latlon and offset latlon are not the same AND is not equal to null
# traj_ors = traj[(traj['latlon'] != traj['latlon_shift']) & (pd.notna(traj['latlon_shift']))]
# De-duplicate
# traj_ors = traj_ors.drop_duplicates(subset=['latlon','latlon_shift'])
traj_ors = traj.copy(deep=True)
```

```python
traj['node_id_d'] = traj.groupby('person_id')['node_id_o'].shift(-1)
traj['node_id_d'] = traj['node_id_d'].fillna(traj['node_id_o'])
traj['node_id_d'] = traj['node_id_d'].astype(int) 
traj
```

#### Calculate the shortest route between each pair of stops

```python
origs = [int(o) for o in traj['node_id_o']]
dests = [int(d) for d in traj['node_id_d']]
```

```python
%%time
distances = transit_ped_net.shortest_path_lengths(origs, dests) # Recall, our original network is set to walking at 3 MPH
# distances
```

```python
routes_list = []

person_id = 1
sequence = 1

for index, row in traj.iterrows():
    # Get shortest path between origin and destination as a list of consecutive nodes
    node_connect = transit_ped_net.shortest_path(row['node_id_o'], row['node_id_d'])
    
    # Retrieve the details of the individual trajectory in focus as a dict with keys as column names and values as individual values
    routes_dict = dict(traj.loc[index])
    
    # Start sequence of travel for the individual trajectory
    if person_id != row['person_id']:
        person_id += 1
        sequence = 1
    
    # If origin is equal to destination, move to the next O-D pair
    if len(node_connect) > 1:
        
        
        for idx, elem in enumerate(node_connect):
            
            next_idx = idx + 1
            sequence += 1
            
            # If the destination node index is less than the length of all intermediate nodes inclusive of and between origin and destination node
            if next_idx < len(node_connect):
                
                if traj.loc[index+1,'travel_type'] != 'stay':
                    travel_type = traj.loc[index+1,'travel_type']
                else:
                    travel_type = 'stay'
                
                routes_dict.update({'sequence': sequence,
                                    'poi_type': 'travel',
                                    'activity_type': travel_type,
                                    'travel_type': travel_type,
                                    'latitude': access_nw.net_nodes.loc[elem, 'y'],
                                    'longitude': access_nw.net_nodes.loc[elem, 'x'],
                                    'datetime': row['datetime'] + timedelta(seconds=sequence),
                                    'node_id_o': elem,
                                    'node_id_d': node_connect[next_idx]})
                
                routes_list.append(routes_dict.copy())
            
            # Else, the last intermediate node is reached
            else:   

                routes_dict.update({'sequence': sequence,
                                    #'travel_type': 'stay',
                                    'latitude': access_nw.net_nodes.loc[elem, 'y'],
                                    'longitude': access_nw.net_nodes.loc[elem, 'x'],
                                    'datetime': row['datetime'] + timedelta(seconds=sequence),
                                    'node_id_o': elem,
                                    'node_id_d': elem})

                routes_list.append(routes_dict.copy())            
    else:
        if sequence != 1:
            sequence += 1
        routes_dict.update({'sequence': sequence,
                            'travel_type': 'stay',
                           })
        routes_list.append(routes_dict)
        
        
routes = pd.DataFrame(routes_list, columns=traj.columns.values)
routes
```
