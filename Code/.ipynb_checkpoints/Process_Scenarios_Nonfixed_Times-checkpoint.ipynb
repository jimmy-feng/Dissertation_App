```python
import pandas as pd
```

#### Identify periods of flexible time in a person's daily life

```python
# Load data
stops = pd.read_csv('../Data/Scenarios_Synthetic_Data.csv', index_col=0)

# Shift the fixed activity type and store as new column
stops['fa_shift'] = stops.groupby('person_id')['fixed_activity'].shift(-1)
# Shift the travel type and store as new column
stops['tt_shift'] = stops.groupby('person_id')['travel_type'].shift(-1)
# Retain only rows in which the current and next immediate points in a person's trajectory are not fixed
nonfixed = stops[(stops['fixed_activity'] == 'n') | (stops['fa_shift'] == 'n')]
```

```python
# Placeholders
nonfixed_list = []
nonfixed_dict = {}
time_available = 0

for idx, row in nonfixed.iterrows():
    
    # Beginning location of non-fixed period
    if (row['fixed_activity'] == 'y') & (row['fa_shift'] == 'n'):
        time_available = 0
        nonfixed_dict['person_id'] = row['person_id']
        nonfixed_dict['latlon_start'] = row['latlon']
        nonfixed_dict['free_time_start'] = row['time_shift']
        if (row['tt_shift'] == 'stay'):
            nonfixed_dict['travel_type'] = stops[(stops['person_id'] == row['person_id']) & (stops['travel_type'] != 'stay')]['travel_type'].mode()[0]
        else:
            nonfixed_dict['travel_type'] = row['tt_shift']
    
    # During non-fixed period
    elif (row['fixed_activity'] == 'n') & (row['fa_shift'] == 'n'):
        # Add duration
        time_available += row['time_available']
    
    # Ending location of non-fixed period
    elif (row['fixed_activity'] == 'n') & (row['fa_shift'] == 'y'):
        nonfixed_dict['latlon_end'] = row['latlon']
        nonfixed_dict['free_time_end'] = row['time_shift']
        
        # Calculate amount of free-time beginning from every person's non-fixed activity locations until their next fixed activity
        time_available += row['time_available']
        nonfixed_dict['time_available'] = time_available
        nonfixed_list.append(nonfixed_dict)
        # Reset dictionary for next set of non-fixed period
        nonfixed_dict = {}

nonfixed_df = pd.DataFrame(nonfixed_list)
nonfixed_df = nonfixed_df[~nonfixed_df['time_available'].isna()]
nonfixed_df
#nonfixed_df.to_csv('../Data/Scenarios_Synthetic_Data_Nonfixed_Times.csv')
```

```python
# Separate latitude and longitude columns for both start and end locations

latlons_list = []
latlons = {}

for idx, row in nonfixed_df.iterrows():
    latlons['lat_start'], latlons['lon_start'] = row['latlon_start'].split(',')
    latlons['lat_end'], latlons['lon_end'] = row['latlon_end'].split(',')
    latlons_list.append(latlons)
    latlons = {}

latlons_df = pd.DataFrame(latlons_list)
nonfixed_df = pd.merge(nonfixed_df, latlons_df, left_index=True, right_index=True)
nonfixed_df.to_csv('../Data/Scenarios_Synthetic_Data_Nonfixed_Times.csv')
nonfixed_df
```
